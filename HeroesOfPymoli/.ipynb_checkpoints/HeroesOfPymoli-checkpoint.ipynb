```python
# Import pd dependency 
import pandas as pd
```

```python
# Define file and path where data are stored to perform analysis 
csv_path = "Resources/purchase_data.csv"
purchase_data_df = pd.read_csv(csv_path)
purchase_data_df.head()
```

```python
# Find the number of SNs (players) to assess the number of active players
player_count = purchase_data_df['SN'].nunique()
players_count = pd.DataFrame({'Total Players': player_count}, index=[0])
players_count
                    
```

```python

# Determine the number of unique purchased items
uniqueitems = purchase_data_df['Item Name'].nunique()

# Determine the average purchase price
averagepurchaseprice = purchase_data_df ['Price'].mean()

# Determine total number of purchases
totalnumberofpurchases = purchase_data_df['Purchase ID'].count()

# Deteremine the total revenue
totalrevenue = purchase_data_df['Price'].sum()

# Create the DataFrame to show a summary of data for the unique purchased items
purchasedatasummary_df = pd.DataFrame([{'Number of Unique Items': uniqueitems, 'Average Price': averagepurchaseprice,
                                  'Number of Purchases': totalnumberofpurchases, 'Total Revenue': totalrevenue}])

# Formatting DataFrame 
purchasedatasummary_df['Average Price'] = purchasedatasummary_df['Average Price'].map('${:.2f}'.format)
purchasedatasummary_df['Total Revenue'] = purchasedatasummary_df['Total Revenue'].map('${:,.2f}'.format)
purchasedatasummary_df
```

```python
# Find the percentage and count of players by gender through grouping

# Defining the data to pass througn by grouping gender
gendercounts_df = purchase_data_df.groupby('Gender')

# Total of unique players by gender
totalbygender = gendercounts_df.nunique()['SN']

# Determine the percentage of unique players by gender  
percentageofplayers = totalbygender/ player_count * 100

# Create Dataframe to display the number of players by gender 
gendercounts = pd.DataFrame({'Total Count': totalbygender, 'Percentage of Players': percentageofplayers, })

# Formatting the results and rounding to two decimal places for the percentage as weel as sort by descending order
gendercounts.sort_values(['Total Count'], ascending = False).style.format({'Percentage of Players':'{:.2f}%'}) 
```

```python
# Analysis by age demographics

# Determine only unique players are counted 
playersbyage = purchase_data_df.loc[:,['Age','SN']]
playersbyage = playersbyage.drop_duplicates()
ageofplayersnodup = playersbyage['Age'].value_counts()

# Create the bins to orgainze player by age
bins_age = [0, 9, 14, 19, 24, 29, 34, 39, 89]

# Age indexing of bins
num = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Creating age grouping from the bins
agedata = pd.cut(playersbyage['Age'], bins_age, labels=num).value_counts()

# Determine the precentage of players by age
agebypercent = agedata / player_count * 100

# Create DataFrame to display the precentage of players by age
agesummary = pd.DataFrame({'Total Count': agedata, 'PERCENTAGE of Players': agebypercent,
})

# Formatting DataFrame to display the precentage of players by age
agesummary ['PERCENTAGE of Players'] = agesummary['PERCENTAGE of Players'].map('{:.2f}%'.format)

#sort by index 
agesummary.sort_index()
```

```python
# Analysis of purchased items by age

# DataFrame to determine purchases by age 
age_df = pd.DataFrame(purchase_data_df)

# Creating age grouping from the bins
agesumm = age_df.groupby(pd.cut(age_df['Age'], bins_age, labels=num))

# Create DataFrame to determine purchases by age group 
agesumm_df = pd.DataFrame({'Purchase Count':agesumm['Price'].count(), 'Average Purchase Price':agesumm['Price'].mean(),
                            'Total Purchase Value':agesumm['Price'].sum(),
                            'Avg Total Purchase per Person':agesumm['Price'].sum()/agesumm['SN'].nunique()})

# Formatting DataFrame 
agesumm_df['Average Purchase Price'] = agesumm_df['Average Purchase Price'].map('${:.2f}'.format)
agesumm_df['Total Purchase Value'] = agesumm_df['Total Purchase Value'].map('${:,.2f}'.format)
agesumm_df['Avg Total Purchase per Person'] = agesumm_df['Avg Total Purchase per Person'].map('${:.2f}'.format)
agesumm_df
```

```python
#Find the total spenders and then display the top five

# Defining the DataFrame to find the items
topspenders_df = pd.DataFrame(purchase_data_df)

#Determine the spending of players by grouping
spenders = topspenders_df.groupby(['SN'])

#Create DataFrame to display the results and sort total purchase value in decending order
spenders_df = pd.DataFrame({'Purchase Count':spenders['Price'].count(), 'Average Purchase Price':spenders['Price'].mean(),
                            'Total Purchase Value':spenders['Price'].sum()}).sort_values('Total Purchase Value', ascending=False)

# Formatting DataFrame
spenders_df['Average Purchase Price'] = spenders_df['Average Purchase Price'].map('${:.2f}'.format)
spenders_df['Total Purchase Value'] = spenders_df['Total Purchase Value'].map('${:.2f}'.format)

# Display the five players who are top spenders
spenders_df.head()
```

```python
# Find the most popular items and display the top five

# Defining the DataFrame to find the items 
mostpopularitems_df = pd.DataFrame(purchase_data_df)

# Determine the most popular items by passing in items through grouping 
mostpopularitems = mostpopularitems_df.groupby(['Item ID', 'Item Name'])

# Create a DataFrame to display results and sort by total purchase value in descending order
mostpopularitems_df = pd.DataFrame({'Purchase Count':mostpopularitems['Price'].count(), 'Item Price':mostpopularitems['Price'].mean(),
                            'Total Purchase Value':mostpopularitems['Price'].sum()}).sort_values('Purchase Count', ascending=False)

# Formatting DataFrame 
mostpopularitems_df['Item Price'] = mostpopularitems_df['Item Price'].map('${:.2f}'.format)
mostpopularitems_df['Total Purchase Value'] = mostpopularitems_df['Total Purchase Value'].map('${:.2f}'.format)

# Display the five most popular items
mostpopularitems_df.head()
```

```python
# Find the most profitable items and display the top five

# Defining the DataFrame to find the items 
mostprofitable_df = pd.DataFrame(purchase_data_df)

# Determine the most profitable items by passing in items through grouping 
mostprofitable = mostprofitable_df.groupby(['Item ID', 'Item Name'])

# Create a DataFrame to dispaly results and sort by total purchase value in descending order
mostprofitable_df = pd.DataFrame({'Purchase Count':mostprofitable['Price'].count(), 'Item Price':mostprofitable['Price'].mean(),
                            'Total Purchase Value':mostprofitable['Price'].sum()}).sort_values('Total Purchase Value', ascending=False)

# Formatting DataFrame
mostprofitable_df['Item Price'] = mostprofitable_df['Item Price'].map('${:.2f}'.format)
mostprofitable_df['Total Purchase Value'] = mostprofitable_df['Total Purchase Value'].map('${:.2f}'.format)

# Display the five most profitable items
mostprofitable_df.head()
```
